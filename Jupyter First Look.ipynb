# Boxplot Example

_An introduction using R_ by Michael J. Crawley, p. 297.

```r
set.seed(11)

plotboxes <- function(N)
{

  a <- rnorm(N, mean=0.00, sd=1)
  b <- rnorm(N, mean=0.00, sd=1)
  c <- rnorm(N, mean=0.50, sd=1)
  d <- rnorm(N, mean=1.00, sd=1)
  e <- rnorm(N, mean=1.50, sd=1)
  f <- rnorm(N, mean=2.00, sd=1)

  boxplot(data.frame(a,b,c,d,e,f),
    notch=TRUE, main=paste("N = ", N))

}

par(mfrow=c(1,2))
plotboxes(50)
plotboxes(100)          
```

# Matrix Color Map Example

```r
set.seed(19)
row.count <- 10
col.count <- 6

m <- round( matrix(rnorm(row.count*col.count),
                   row.count, col.count), 4)

x <- seq(from=0, to=1, length=ncol(m))
y <- seq(from=0, to=1, length=nrow(m))


ColorRamp <- rgb( seq(0,1,length=256),  # Red
                  seq(0,1,length=256),  # Green
                  seq(1,0,length=256))  # Blue

image(t(m), col=ColorRamp, axes=FALSE)

for (i in row.count:1)
{
    text(x,y[i],labels=sprintf("%.4f", m[i,]))
}

axis(1, at=seq(0,1,length.out=ncol(m)),
        labels=1:ncol(m), tick=TRUE)
axis(2, at=seq(0,1,length.out=nrow(m)),
        labels=1:nrow(m), tick=TRUE, las=2)

m                           
```
